### Rolling

````python
import pandas_datareader.data as web  ### permite obtener data frames directamente de internet

gs = web.DataReader("GS", data_source='robinhood', start='2016-01-01',
                    end='2017-01-01')

gs.head()
````

````python
print "memoria total: ",gs.memory_usage().sum() ,"bytes"
````

````python
del(gs['session'])
del(gs['interpolated'])
````

````python
gs.index.names=['stock','Date']

gs.head()
````

````python
gs[['close_price','high_price']] = gs[['close_price','high_price']].apply(pd.to_numeric)
````

````python
gs.shape
````

````python
media1=gs.close_price.rolling(10).mean()
print media1[:20]
````

````python
gs.close_price.rolling(10).sum()[:20]
````

````python
media2=gs.close_price.rolling(window=10, center=True).mean()
print media2[:20]
````

````python
media1.plot(label='Ventana=10, Centro=F')
media2.plot(label='Ventana=10, Centro=V')
gs.close_price.plot(label='Precio de cierre')

plt.legend()
````

````python
gs.close_price.rolling(window=10, center=True).std().plot()
````

````python
gs.close_price.rolling(10).corr(gs.high_price).plot()
````

````python
gs.close_price.rolling(10).corr(gs.high_price).mean()
````
